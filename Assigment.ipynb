```python
# Import necessary library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Questions
1. Load the 'employee_data.csv' file into Python for analysis.

```python
df = pd.read_csv('/content/employee_data.csv')
df.head(20)
```

Questions
2. Handle missing values in the 'salary', 'performance_score', and 'education' columns.

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Fill missing values in 'salary' with the mean
df['salary'].fillna(df['salary'].mean(), inplace=True)

# Fill missing values in 'performance_score' with the median
df['performance_score'].fillna(df['performance_score'].median(), inplace=True)
```

```python
# Fill missing values in 'education' with the mode
df['education'].fillna(df['education'].mode()[0], inplace=True)
```

```python
# Check for missing values again
print(df.isnull().sum())
```

Questions
3. Correct any inconsistencies in the 'gender' column.

```python
# Check unique values in 'gender' column
print(df['gender'].unique())
```

```python
# Standardize gender values to 'Male' and 'Female'
df['gender'] = df['gender'].replace({'male': 'Male', 'female': 'Female', 'M': 'Male', 'F': 'Female'})

# Verify the changes
print(df['gender'].unique())
```

Questions
4. Perform any 4 visualizations which you feel are good for your analysis.

```python
#1. Distribution of Salaries
# Histogram of salaries
plt.figure(figsize=(8, 6))
sns.histplot(df['salary'], kde=True)
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Count')
plt.show()
```

The salary range is from 40,000 to 140,000. The bars indicate the count of individuals within specific salary ranges. The highest count seen at the 40,000  and  140,000 salary ranges.

```python
#2. Performance Score by Department
# Box plot of performance scores by department
plt.figure(figsize=(10, 6))
sns.boxplot(x='department', y='performance_score', data=df)
plt.title('Performance Score by Department')
plt.xlabel('Department')
plt.ylabel('Performance Score')
plt.xticks(rotation=45)
plt.show()
```

The box plots show that the performance scores across departments have a similar spread, with most medians around 3.0 to 3.5 and the scores generally ranging between 1.0 to 5.0.

```python
#3. Gender Distribution across Departments
# Count plot of gender distribution across departments
plt.figure(figsize=(10, 6))
sns.countplot(x='department', hue='gender', data=df)
plt.title('Gender Distribution across Departments')
plt.xlabel('Department')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#4. Correlation Heatmap
# Correlation heatmap of numerical variables
plt.figure(figsize=(8, 6))
# Select only numerical columns for correlation calculation
numerical_df = df.select_dtypes(include=['float', 'int'])
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
